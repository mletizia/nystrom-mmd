## Higgs

```python
#fullrank_time_pow_nft = load_results("/home/marcol/NystromMMD_new/output_higgs_paper/", method='fullrank')
uniform_time_pow_nft = load_results_CG("/home/marcol/NystromMMD_new/output_CG/", method='uniform')
rlss_time_pow_nft = load_results_CG("/home/marcol/NystromMMD_new/output_CG/", method='rlss')
rff_time_pow_nft = load_results_CG("/home/marcol/NystromMMD_new/output_CG/", method='rff')
```

```python
rho = [0.51, 0.54, 0.57, 0.60, 0.63, 0.66]

nfeat = 4

#fullrank_power = np.asarray([power_interval(el, 400) for el in fullrank_time_pow_nft[:,1]])
uniform_powers = np.asarray([power_interval(el, 400) for el in uniform_time_pow_nft[:,1, nfeat]])
rff_powers = np.asarray([power_interval(el, 400) for el in rff_time_pow_nft[:,1, nfeat]])
rlss_powers = np.asarray([power_interval(el, 400) for el in rlss_time_pow_nft[:,1, nfeat]])

plt.figure(figsize=(8, 5))
plt.plot(rho, uniform_powers[:,0], '-v', markersize=8, label=r'Nyström-uniform (ours, $\ell=\sqrt{n}$)', c='#377eb8')
plt.fill_between(rho, 
                 uniform_powers[:,1], 
                 uniform_powers[:,2], 
                 alpha=0.2, color='#377eb8')
plt.plot(rho, rlss_powers[:,0], '-p', markersize=8, label=r'Nyström-AKRLS (ours, $\ell=\sqrt{n}$)', c='#ff7f00')
plt.fill_between(rho, 
                 rlss_powers[:,1], 
                 rlss_powers[:,2], 
                 alpha=0.2, color='#ff7f00')
plt.plot(rho, rff_powers[:,0], '-x', markersize=8, label=r'RFF  ($\ell=\sqrt{n}$)', c='#4daf4a')
plt.fill_between(rho, 
                 rff_powers[:,1], 
                 rff_powers[:,2], 
                 alpha=0.2, color='#4daf4a')
# plt.plot(fullrank_time_pow_nft[:,-1], fullrank_power[:,0], '-o', markersize=8, label=r'Full rank', c='#984ea3')
# plt.fill_between(fullrank_time_pow_nft[:,-1], 
#                  fullrank_power[:,1], 
#                  fullrank_power[:,2], 
#                  alpha=0.5, color='#984ea3')
plt.ylabel(r'Power ($\alpha=0.05$)', fontsize=20)
plt.xlabel(r'$\rho_2$', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.locator_params(nbins=6, axis='x')
#plt.xscale('log')
plt.legend(loc=2, fontsize=15)
#plt.title("Correlated Gaussians", fontsize=20)
plt.xlim(0.5,0.67)
plt.grid()
plt.show()
```
